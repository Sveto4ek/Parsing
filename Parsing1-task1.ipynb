```python
import requests
```

```python
import os
import json
from pprint import pprint
service = 'https://api.github.com/users/Sveto4ek/repos'
username = 'Sveto4ek'
token = os.environ.get("GITHUB_TOKEN")
repos = requests.get(service, auth=(username, token))
#for repo in repos.json(): 
#  if not repo['private']: print(f"{repo['name']}: {repo['html_url']}")
r_list = {}
for repo in repos.json(): 
  if not repo['private']: r_list[repo['name']] = repo['html_url']
pprint(r_list)
```

```python
with open('repos.json', 'w') as rps:
  json.dump(r_list, rps)
```

```python
with open('repos.json', 'r', encoding='utf-8') as f:
    text = json.load(f)
    pprint(text) 
```
